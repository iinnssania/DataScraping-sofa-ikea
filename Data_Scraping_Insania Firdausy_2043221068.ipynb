```python
import pandas as pd
from bs4  import BeautifulSoup

import chromedriver_autoinstaller
from selenium import webdriver
```

```python
chromedriver_autoinstaller.install()

browser = webdriver.Chrome()
```

```python
browser.get('https://www.ikea.co.id/in/produk/sofa-dan-kursi-berlengan')
```

```python
soup = BeautifulSoup(browser.page_source, 'lxml')
```

```python
soup
```

```python
soup.find_all('div', class_='itemInfo')
```

```python
products = soup.find_all('div', class_='itemInfo')
```

```python
data = []
for product in products :
    Nama = product.find(class_='display-7').text.strip()
    Detail = product.find(class_='itemFacts').text.strip()
    Harga = product.find(class_='itemPrice-wrapper').text.strip()
    
    product = {
        'Nama': Nama, 
        'Detail': Detail,
        'Harga': Harga,
    }
    
    data.append(product)
```

```python
pd.DataFrame(data)
```

```python
pd.DataFrame(data).to_excel('Data__Sofa_Ikea.xlsx')
```
